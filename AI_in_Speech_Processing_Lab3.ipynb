#A1. Use librosa.effects.trim() to remove the silence parts of speech from beginning and 
end of your recorded signal. Listen to the new signal and perceptually compare the audio with 
original. 

```python
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import IPython.display as ipd

audio = r"E:\AI in Speech\Lab\Lab3\audio.wav"
y, sr = librosa.load(audio, sr=None)
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
trimmed_audio = "trimmed_audio.wav"
sf.write(trimmed_audio, y_trimmed, sr)


print("Original Audio:")
ipd.display(ipd.Audio(audio))

print("Trimmed Audio:")
ipd.display(ipd.Audio(trimmed_audio))


plt.figure(figsize=(12, 6))


plt.subplot(2, 1, 1)
plt.plot(y)
plt.title('Original Audio')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(y_trimmed)
plt.title('Trimmed Audio')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

#A2. Use librosa.effects.split() to split the recorded speech with detected silences. Play 
around with the top_db parameter and see the effects of split. Listen to the generated signals and 
observe the split quality. 

```python
import librosa
import soundfile as sf
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt

audio_file = r"E:\AI in Speech\Lab\Lab3\audio.wav"
y, sr = librosa.load(audio_file, sr=None)

top_db_values = [10, 20, 30]
for top_db in top_db_values:
    intervals = librosa.effects.split(y, top_db=top_db)
    split_audio = np.concatenate([y[start:end] for start, end in intervals])
    split_audio_file = f"split_audio_top_db_{top_db}.wav"
    sf.write(split_audio_file, split_audio, sr)
    print(f"Number of segments with top_db={top_db}: {len(intervals)}")
    print(f"Split audio with top_db={top_db}:")
    display(Audio(split_audio_file))


top_db_values = [10, 20, 30]
for top_db in top_db_values:
    intervals = librosa.effects.split(y, top_db=top_db)
    split_audio = np.concatenate([y[start:end] for start, end in intervals])
    plt.figure(figsize=(10, 3))
    plt.plot(split_audio)
    plt.title(f"Split audio with top_db={top_db}")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.show()
```
